```python
import pandas as pd
import numpy as np
import numpy
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
import seaborn as sns


# Department of Environmental Protection (DEP) dataset on 
# annual population of NYC and it's water consumption

# read data from csv file using pandas
df=pd.read_csv('Water_Consumption_in_the_City_of_New_York.csv',delimiter=',')
```

```python
df.head()
```

```python
# assign variable to columns of interest
x = df['New York City Population']
y= df['NYC Consumption(Million gallons per day)']

# create scatter plot with variables 
plt.figure(figsize=(8, 8))
plt.scatter(x, y)
plt.title('Correlation between NYC Water Consumption and Population', fontsize=20)
plt.xlabel('Population(millions)', fontsize=15)
plt.ylabel('Water Consumption (million gallons per day)', fontsize=15)
# plot line of best fit
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='yellow')
plt.show()





```

```python
# calculate the coefficient of determination
corr_matrix = numpy.corrcoef(x, y)
corr = corr_matrix[0,1]
R_sq = corr**2

print(R_sq)

# These results saying 89% of the points should fall within the regression line
```

```python
# use scipy function to calculate the slope of the regression line, correlation coefficient, and more....
linregress(x, y)
```
